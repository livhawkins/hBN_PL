```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Measurement:
    '''A class to store and analyze measurement data.
    
    Args:
        name (str): Name of the measurement.
        units (str): Units of the measured values.
    '''

    def __init__(self, name: str, units: str) -> None:

        '''Initialize the Measurement class.
        
        Args:
            name (str): Name of the measurement.
            units (str): Units of the measured values.
        '''

        print("measurement class initialized")
        self.name = name
        self.units = units
        self.times = []
        self.values = []
        self.offset = 0

    def add_point(self, t, x):
        '''Add a data point to the measurement.
        
        Args:
            t (float): Time of the measurement.
            x (float): Measured value.
        '''
        self.times.append(t + self.offset)
        self.values.append(x)

    def mean(self):
        return np.mean(self.values)
    
    def plot(self):
        plt.plot(self.times, self.values)
        plt.xlabel("Time (s)")
        plt.ylabel(f"Value ({self.units})")
        plt.title(self.name)
        plt.show()

    def calibrate(self, offset):
        self.offset += offset

    def __repr__(self): 
        return f"Measurement(name={self.name}, units={self.units}, points={len(self.times)})"
```

```python
measurement_1 = Measurement(name = "tara's measurement", units = "volts")
```

```python
measurement_1.times, measurement_1.values
```

```python
measurement_1.plot()
```
