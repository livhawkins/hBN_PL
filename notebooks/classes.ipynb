Classes - an object where you can store things and also perform functions to the things that are stored

```python
import numpy as numpy
import matplotlib.pyplot as plt
```

```python
class Measurement:
    def __init__(self, name, units):
        print("Measurement class initialised")
        self.name = name
        self.units = units
        self.times = []
        self.values = []
        self.offsets = 0


    def add_point(self, t, x):
        self.times.append(t)
        self.values.append(x)
```

```python
measurement_1 = Measurement(name = "tara's measurement",
                             units = "volts")
```

```python
measurement_1.name #think of measurement_1 as self inside the class
measurement_1.units
```

```python
measurement_1.add_point(0,1.2)
```
